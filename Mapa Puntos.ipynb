```python
import streamlit as st
import pandas as pd
import folium
from streamlit_folium import folium_static
from geopy.distance import geodesic

# Leer los datos del archivo Excel
@st.cache(allow_output_mutation=True)
def load_data(filename):
    return pd.read_excel(filename)

df = load_data('Data/Maestra_Puntos_Reciclaje.xlsx')

# Obtener la ubicación del usuario (aquí uso una ubicación ficticia como ejemplo)
# En un caso real, podrías pedir esta información al usuario o intentar obtenerla automáticamente
user_location = [-33.6479903, -70.7096326] # Ejemplo: Latitud y Longitud de Nueva York

# Crear un mapa con Folium
m = folium.Map(location=user_location, zoom_start=12)

# Añadir un marcador para la ubicación del usuario
folium.Marker(user_location, popup='Tu Ubicación', icon=folium.Icon(color='red')).add_to(m)

# Añadir marcadores para los puntos de reciclaje
for index, row in df.iterrows():
    folium.Marker([row['Latitud'], row['Longitud']], popup=row['Nombre']).add_to(m)

# Mostrar el mapa en Streamlit
folium_static(m)
```
